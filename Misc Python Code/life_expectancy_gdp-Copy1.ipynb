```python
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', hue=data.Country, data=data)
plt.show()
plt.clf()
```

```python
trans_GDP=np.log(data['Life expectancy at birth (years)'])
```

```python
print(trans_GDP.head())
```

```python
data['GDP']=trans_GDP
```

```python
print(data.head())
```

```python
sns.scatterplot(x='GDP', y='Life expectancy at birth (years)', hue=data.Country, data=data, x_jitter=0.30, y_jitter= 0.30 )
plt.show()
plt.clf()
```
